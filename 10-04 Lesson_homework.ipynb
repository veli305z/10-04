#### Написать функцию, которая производит поиск и выводит на экран вакансии с заработной платой больше введённой суммы (необходимо анализировать оба поля зарплаты).

```python
import requests
import json
from pprint import pprint
```

```python
url = 'https://api.hh.ru/vacancies' # нашел в интернете инфо, что hh можно парсить через api

params = {
        'text': 'NAME:pithon', # Текст фильтра. В имени должно быть слово "pithon"
        'per_page': 100 # Кол-во вакансий на 1 странице
    }


response = requests.get(url, params=params)
```

```python
pprint(response.status_code)
```

```python
pprint(response.json())
```

```python
a = len(response.json()['items'])
```

```python
value = input("Введите зарплату, на которую претендуете: ")
```

```python
i=0
value = int(value)
```

Если зарплата не определена, вакансию пропускаем и проклинаем работодателя :)

Если верхняя граница з/п определена, то "хотелки" меньше нее подходят под условие

Если верхняя граница з/п не определена, а нижняя определена,  "хотелки" меньше нижней границы подходят под условие

```python
while i < a:
    if response.json()['items'][i]['salary'] == None:
        salary_min = 0
        salary_max = 0
    else:
        salary_min = int(response.json()['items'][i]['salary']['from'] or 0) # обработка ошибки про 'NonType' объекта
        salary_max = int(response.json()['items'][i]['salary']['to'] or 0)
        if value <= salary_max:
                print(f"Вам подходит вакансия № {response.json()['items'][i]['id']} : {response.json()['items'][i]['name']} с зарплатой от {salary_min} до {salary_max}")
        if value <= salary_min:
            print(f"Вам подходит вакансия № {response.json()['items'][i]['id']} : {response.json()['items'][i]['name']} с зарплатой от {salary_min} до {salary_max}")
    i+=1
```
